<h1 align = "center">Parameter Visualization</h1>

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
import seaborn as sns
from tqdm import tqdm

import latte
import latte.nn as nn
import latte.optim as optim
import latte.functional as F
import latte.utils.data as data

import lattevision as lv
import lattevision.datasets as dsets
import lattevision.transforms as T
```

```python
class Model(nn.Module):
    def __init__(self) -> None:
        super().__init__()
        self.fc1 = nn.Linear(784, 512)
        self.relu = nn.ReLU()
        self.fc2 = nn.Linear(512, 10)

    def forward(self, x):
        out = self.fc1(x)
        out = self.relu(out)
        out = self.fc2(out)
        return out
```

```python
model = Model()
model.load("./best_models/best_model.npy")

fc1_weights = model.fc1.weight.data
fc1_shape = model.fc1.weight.shape
fc2_weights = model.fc2.weight.data
fc2_shape = model.fc2.weight.shape

plt.figure(figsize=(18, 12))
gs = GridSpec(6, 1)

plt.subplot(gs[:5, :])
sns.heatmap(fc1_weights.T, cbar=True, square=True)
plt.title(f"FC1 Weights (size: {fc1_shape})")
plt.axis('off')

plt.subplot(gs[-1:, :])
sns.heatmap(fc2_weights.T, cbar=True, square=True)
plt.title(f"FC2 Weights (size: {fc2_shape})")
plt.axis('off')

plt.show()
```
